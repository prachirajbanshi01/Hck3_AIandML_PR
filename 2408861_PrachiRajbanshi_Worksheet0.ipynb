```python
# 4.1 Exercise on Functions
# Task 1

def convert_units(value, conversion_type):

  #Conversion Constants
  M_TO_FT = 3.28084
  KG_TO_LBS = 2.20462
  L_TO_GAL = 0.26417

  if conversion_type == 1:
    return value * M_TO_FT, "Feet"
  elif conversion_type == 2:
    return value / M_TO_FT, "Meters"
  elif conversion_type == 3:
    return value * KG_TO_LBS, "Pounds"
  elif conversion_type == 4:
    return value / KG_TO_LBS, "Kilograms"
  elif conversion_type == 5:
    return value * L_TO_GAL, "Gallons"
  elif conversion_type == 6:
    return value / L_TO_GAL, "Liters"
  else:
    return None

def main():
  print("=====Multi-Unit Converter=====")
  print("==============================")
  print("Select a conversion: ")
  print("1: Meters to Feet | 2: Feet to Meters")
  print("3: Kilograms to Pounds | 4: Pounds to Kilograms")
  print("5: Liters to Gallons | 6: Gallons to Liters")

  try:
    choice = int(input("\nEnter the number of your choice (1-6): "))

    if choice not in range(1, 7):
      print("Error: Please select a valid option between 1 to 6")
      return

    val = float(input("Enter the value to convert: "))

    result, unit = convert_units(val, choice)

    if result is not None:
      print(f"\nResult: {val} converts to {result} {unit}")

  except ValueError:
    print("Invalid Input: Please enter numeric values only")

main()


```

```python
# 4.3 Exercise on Nested List
# Task 2 - Accessing Nested List Elements

def access_nested_element(lst, indices):
  current_element = lst

  try:
    for index in indices:
      current_element = current_element[index]
    return current_element
  except (IndexError, TypeError):
    return "Error: Invalid index path or element is not a list"

nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
#path = [0,1]
result = access_nested_element(nested_list, [1,2])

print(f"Result: {result}")
```

```python
# 4.3 Exercise on Nested List
# Task 3 - Sum of All Elements in a Nested List

def sum_nested(lst):
  total = 0
  for item in lst:
    if isinstance(item, list):
      total += sum_nested(item)
    else:
      total += item
  return total

nested_data = [1, [2, 3], [4, [5, 6]], 7]
result = sum_nested(nested_data)

print(f"Sum: {result}")
```

```python
# 4.3 Exercise on Nested List
# Task 5 - Find the Maximum Element in a Nested List

def find_max(lst):
  current_max = None

  for item in lst:
    if isinstance(item, list):
      temp_max = find_max(item)
    else:
      temp_max = item

    if temp_max is not None:
      if current_max is None or temp_max > current_max:
        current_max = temp_max
  return current_max

input_list = [[1, 2], [3, 2], [4, 5]]
print(f"The maximum eement is: {find_max(input_list)}")
```

```python
# 4.3 Exercise on Nested List
# Task 6 - Count Occurence of an Element in a Nested List

def count_occurrences(lst, elem):
  count = 0
  for item in lst:
    if isinstance(item, list):
      count += count_occurrences(item, elem)
    elif item == elem:
      count += 1
  return count

input_list = [[1, 2], [3, 4], [4,4]]
target = 4

result = count_occurrences(input_list, target)
print(f"The number of occurences of {target} is: {result}")
```

```python
# 4.3 Exercise on Nested List
# Task 7 - Flatten a List of Lists of Lists

def deep_flatten(lst):
    """
    Recursively flattens a list with any depth of nesting.

    Parameters:
    lst (list): The nested list to flatten.

    Returns:
    list: A single-level list containing all elements.
    """
    flat_list = []

    for item in lst:
        if isinstance(item, list):
            # If the item is a list, flatten it and add the results to our list
            flat_list.extend(deep_flatten(item))
        else:
            # If it's a single value, just append it
            flat_list.append(item)

    return flat_list

# Example usage:
input_data = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(input_data)

print(f"Original: {input_data}")
print(f"Flattened: {result}")
# Output: [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
# 4.3 Exercise on Nested List
# Task 8 - Nested List Average

def average_nested(lst):
    """
    Calculates the average of all numeric elements in a nested list.

    Parameters:
    lst (list): A nested list containing numbers.

    Returns:
    float: The average of all elements.
    """
    def get_stats(data):
        total_sum = 0
        total_count = 0

        for item in data:
            if isinstance(item, list):
                # Recurse and add the results to current stats
                s, c = get_stats(item)
                total_sum += s
                total_count += c
            else:
                total_sum += item
                total_count += 1
        return total_sum, total_count

    total_sum, total_count = get_stats(lst)

    # Handle the case of an empty list to avoid ZeroDivisionError
    if total_count == 0:
        return 0

    return total_sum / total_count

# Example usage:
input_data = [[1, 2], [3, 4], [5, 6]]
result = average_nested(input_data)

print(f"The average is: {result}")
```

```python
# Problem 4.1 : Matrix Operations

import numpy as np

# Define matrices A and B
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# --- 1. Prove A * A^-1 = I ---
A_inv = np.linalg.inv(A)
result_1 = np.dot(A, A_inv)
# We use allclose because floating point math might result in tiny values like 1e-16 instead of 0
proof_1 = np.allclose(result_1, np.eye(2))

# --- 2. Prove AB != BA ---
AB = np.dot(A, B)
BA = np.dot(B, A)
proof_2 = not np.array_equal(AB, BA)

# --- 3. Prove (AB)^T = B^T * A^T ---
AB_transpose = AB.T
BT_AT = np.dot(B.T, A.T)
proof_3 = np.allclose(AB_transpose, BT_AT)

# --- Output Results ---
print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
print("-" * 30)
print("1. A * A^-1 Result:\n", result_1)
print(f"   Is it Identity? {proof_1}")
print("-" * 30)
print("2. AB:\n", AB)
print("   BA:\n", BA)
print(f"   Is AB != BA? {proof_2}")
print("-" * 30)
print("3. (AB)^T:\n", AB_transpose)
print("   B^T * A^T:\n", BT_AT)
print(f"   Is (AB)^T == B^T * A^T? {proof_3}")

```
